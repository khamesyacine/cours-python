## 2.1 Calculs en biologie

**Microbiologie** â€” Croissance bactÃ©rienne

````python
N0 = 500                           # population initiale
generations = 5
Nf = N0 * (2 ** generations)       # N = N0 Ã— 2^n
print(f"AprÃ¨s {generations} divisions : {Nf} bactÃ©ries")
````

**Biochimie** â€” Dilution (C1V1 = C2V2)

````python
C1 = 10.0                          # concentration initiale (mol/L)
V1 = 2.0                           # volume prÃ©levÃ© (mL)
V2 = 50.0                          # volume final (mL)
C2 = (C1 * V1) / V2
print(f"Concentration finale : {C2} mol/L")
````

**Immunologie** â€” Taux de sÃ©ropositivitÃ©

````python
positifs = 35
total = 200
taux = (positifs / total) * 100
print(f"SÃ©ropositivitÃ© : {taux:.1f} %")
````

**SantÃ©** â€” Indice de Masse Corporelle (IMC)

````python
poids = 85.0                       # kg
taille = 1.70                      # m
IMC = poids / (taille ** 2)
print(f"IMC = {IMC:.1f} kg/mÂ²")
````

**Nutrition** â€” Besoin calorique journalier (formule simplifiÃ©e)

````python
poids = 70                         # kg
facteur_activite = 1.55            # activitÃ© modÃ©rÃ©e
metabolisme_base = 24 * poids      # kcal/jour
besoin = metabolisme_base * facteur_activite
print(f"Besoin journalier : {besoin:.0f} kcal")
````

**DiabÃ¨te** â€” Dose d'insuline

````python
glycemie_actuelle = 2.80           # g/L
glycemie_cible = 1.00              # g/L
sensibilite = 0.50                 # g/L par unitÃ© d'insuline
dose = (glycemie_actuelle - glycemie_cible) / sensibilite
print(f"Dose d'insuline : {dose:.0f} unitÃ©s")
````

**Cancer** â€” Taux de survie

````python
patients_total = 500
survivants_5ans = 410
taux_survie = (survivants_5ans / patients_total) * 100
print(f"Taux de survie Ã  5 ans : {taux_survie} %")
````

## 3.1 OpÃ©rateurs de comparaison

## 3.2 OpÃ©rateurs logiques : `and`, `or`, `not`

| OpÃ©rateur | Signification |
|-----------|---------------|
| `and` | les **deux** vrais |
| `or` | **au moins un** vrai |
| `not` | **inverse** |

````python
print(True and True)               # â†’ True
print(True and False)              # â†’ False
````

````python
print(True or False)               # â†’ True
print(False or False)              # â†’ False
````

````python
print(not True)                    # â†’ False
print(not False)                   # â†’ True
````

## 3.3 Applications

**Biochimie** â€” pH sanguin normal ? (entre 7.35 et 7.45)

````python
pH = 7.38
normal = pH >= 7.35 and pH <= 7.45
print(f"pH = {pH} â†’ Normal : {normal}")  # â†’ True
````

**DiabÃ¨te** â€” GlycÃ©mie Ã  jeun normale ? (entre 0.7 et 1.1 g/L)

````python
glycemie = 0.95
normal = glycemie >= 0.7 and glycemie <= 1.1
print(f"GlycÃ©mie = {glycemie} â†’ Normal : {normal}")  # â†’ True
````

**Microbiologie** â€” BactÃ©rie multirÃ©sistante ? (rÃ©sistante Ã  au moins 2 antibiotiques)

````python
resist_ampicilline = True
resist_vancomycine = True
multi_resistant = resist_ampicilline and resist_vancomycine
print(f"MultirÃ©sistant : {multi_resistant}")  # â†’ True
````

**Immunologie** â€” Faut-il vacciner ? (pas immunisÃ© OU immunodÃ©primÃ©)

````python
immunise = False
immunodeprime = True
vacciner = not immunise or immunodeprime
print(f"Vacciner : {vacciner}")    # â†’ True
````

**Cancer** â€” Patient Ã  risque ? (fumeur ET antÃ©cÃ©dents familiaux)

````python
fumeur = True
antecedents_familiaux = False
haut_risque = fumeur and antecedents_familiaux
print(f"Haut risque : {haut_risque}")  # â†’ False
````

**SantÃ©** â€” Alerte tension artÃ©rielle ? (systolique > 140 OU diastolique > 90)

````python
systolique = 145
diastolique = 85
hypertension = systolique > 140 or diastolique > 90
print(f"Hypertension : {hypertension}")  # â†’ True
````

**Nutrition** â€” Carence en fer ? (hÃ©moglobine basse ET ferritine basse)

````python
hemoglobine = 10.5                 # g/dL (normal > 12)
ferritine = 8.0                    # ng/mL (normal > 20)
carence_fer = hemoglobine < 12 and ferritine < 20
print(f"Carence en fer : {carence_fer}")  # â†’ True
````
