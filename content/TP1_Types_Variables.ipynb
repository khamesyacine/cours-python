## 1. Les entiers (`int`)

```python
age = 25                   # on crée une variable entière
print(age)                  # affiche : 25
```

```python
type(age)                  # vérifie le type → int
```

```python
nb_poissons = 120          # nombre de poissons comptés
nb_sites = 4               # nombre de sites d'étude
```

```python
total = nb_poissons + 30   # addition → 150
print(total)
```

```python
double = nb_sites * 2      # multiplication → 8
print(double)
```

```python
par_site = nb_poissons // nb_sites   # division entière → 30
print(par_site)
```

```python
reste = nb_poissons % nb_sites       # reste de la division → 0
print(reste)
```

```python
carre = 5 ** 2             # puissance → 25
print(carre)
```

---
## 2. Les décimaux (`float`)

```python
temperature = 18.5         # nombre à virgule
print(temperature)
```

```python
type(temperature)          # vérifie le type → float
```

```python
pH = 8.2                   # pH de l'eau de mer
salinite = 37.5            # salinité en PSU
```

```python
somme = pH + salinite      # addition de floats → 45.7
print(somme)
```

```python
moyenne = (18.5 + 20.1 + 19.3) / 3   # calcul de moyenne
print(moyenne)
```

```python
# Attention : int + float → float
x = 10 + 2.5              # résultat = 12.5 (float)
print(x, type(x))
```

```python
print(round(moyenne, 2))   # arrondir à 2 décimales
```

## 3. Les chaînes de caractères (`str`)

```python
nom = "Posidonia"          # guillemets doubles
print(nom)
```

```python
espece = 'oceanica'        # guillemets simples (pareil)
print(espece)
```

```python
type(nom)                  # vérifie le type → str
```

```python
nom_complet = nom + " " + espece   # concaténation
print(nom_complet)                   # → Posidonia oceanica
```

```python
print(len(nom_complet))    # longueur de la chaîne → 19
```

```python
print(nom_complet.upper()) # tout en majuscules
print(nom_complet.lower()) # tout en minuscules
```

```python
# f-string : insérer des variables dans du texte
print(f"L'espèce {nom_complet} vit à {temperature}°C")
```

---
## 4. Conversions entre types

```python
x = "42"                   # c'est un texte, pas un nombre !
print(type(x))             # → str
```

```python
x = int("42")              # convertir texte → entier
print(x + 10)              # → 52
```

```python
y = float("18.5")          # convertir texte → décimal
print(y + 1)               # → 19.5
```

```python
z = str(100)               # convertir nombre → texte
print("Il y a " + z + " espèces")   # concaténation OK
```

```python
n = int(7.9)               # float → int (tronque, pas d'arrondi !)
print(n)                   # → 7
```
