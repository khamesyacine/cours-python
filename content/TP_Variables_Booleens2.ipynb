## 1.1 Les types fondamentaux

Python poss√®de plusieurs types de donn√©es de base :

| Type | Nom Python | Exemple | Usage en biologie |
|------|-----------|---------|--------------------|
| Entier | `int` | `42` | Nombre d'individus, d'esp√®ces |
| D√©cimal | `float` | `3.14` | Temp√©rature, pH, concentration |
| Cha√Æne | `str` | `"ADN"` | Noms d'esp√®ces, s√©quences |

| NoneType | `NoneType` | `None` | Donn√©e manquante |

### üî¨ D√©monstration : Identifier les types

````python
# --- Entier (int) ---
nb_especes = 47
print(f"nb_especes = {nb_especes}")
print(f"  Type : {type(nb_especes)}")

# --- D√©cimal (float) ---
temperature = 18.5
print(f"\ntemperature = {temperature}")
print(f"  Type : {type(temperature)}")

# --- Cha√Æne de caract√®res (str) ---
espece = "Posidonia oceanica"
print(f"\nespece = '{espece}'")
print(f"  Type : {type(espece)}")

# --- Bool√©en (bool) ---
est_protegee = True
print(f"\nest_protegee = {est_protegee}")
print(f"  Type : {type(est_protegee)}")

# --- Liste (list) ---
mesures = [23.1, 25.4, 22.8, 27.3, 24.6]
print(f"\nmesures = {mesures}")
print(f"  Type : {type(mesures)}")

# --- Dictionnaire (dict) ---
echantillon = {"site": "Sablettes", "pH": 8.1, "salinite": 37.2}
print(f"\nechantillon = {echantillon}")
print(f"  Type : {type(echantillon)}")

# --- None ---
donnee_manquante = None
print(f"\ndonnee_manquante = {donnee_manquante}")
print(f"  Type : {type(donnee_manquante)}")
````

### üîé La fonction `type()` ‚Äî Votre outil de diagnostic

La fonction `type()` retourne le type de n'importe quelle valeur. C'est l'outil le plus important pour comprendre comment Python interpr√®te vos donn√©es.

> ‚ö†Ô∏è **Pi√®ge courant** : `"42"` (entre guillemets) est une **cha√Æne**, pas un entier !

````python
# Attention √† la diff√©rence !
a = 42       # entier
b = 42.0     # d√©cimal
c = "42"     # cha√Æne de caract√®res !

print(f"a = {a:<8} ‚Üí type = {type(a).__name__}")
print(f"b = {b:<8} ‚Üí type = {type(b).__name__}")
print(f"c = '{c}'    ‚Üí type = {type(c).__name__}")

print(f"\na == b ? {a == b}")   # True (m√™me valeur num√©rique)
print(f"a == c ? {a == c}")     # False (types diff√©rents !)
````

## 2.3 Conversion de types (casting)

Il est souvent n√©cessaire de **convertir** une variable d'un type √† un autre :

| Fonction | Conversion vers | Exemple |
|----------|----------------|----------|
| `int()` | Entier | `int(3.7)` ‚Üí `3` |
| `float()` | D√©cimal | `float("18.5")` ‚Üí `18.5` |
| `str()` | Cha√Æne | `str(42)` ‚Üí `"42"` |
| `bool()` | Bool√©en | `bool(0)` ‚Üí `False` |
| `list()` | Liste | `list("ADN")` ‚Üí `['A','D','N']` |

````python
# Conversions courantes

# str ‚Üí float (ex: lecture d'un fichier CSV)
valeur_texte = "23.7"
valeur_num = float(valeur_texte)
print(f"'{valeur_texte}' (str) ‚Üí {valeur_num} (float)")
print(f"  Calcul possible : {valeur_num} + 10 = {valeur_num + 10}")

# float ‚Üí int (troncature, pas d'arrondi !)
temperature = 18.7
temp_entier = int(temperature)
print(f"\n{temperature} (float) ‚Üí {temp_entier} (int)  ‚ö†Ô∏è troncature !")
print(f"  Pour arrondir : round({temperature}) = {round(temperature)}")

# int ‚Üí str (pour l'affichage)
nb = 42
message = "Il y a " + str(nb) + " esp√®ces"  # sans str() ‚Üí erreur !
print(f"\n{message}")

# Cha√Æne ‚Üí Liste de caract√®res
sequence = "ATCGGATC"
bases = list(sequence)
print(f"\n'{sequence}' ‚Üí {bases}")
print(f"  Nombre de bases : {len(bases)}")
````

### ‚ö†Ô∏è Erreurs de conversion courantes

````python
# Attention aux conversions impossibles !

# Ceci fonctionne :
print(float("3.14"))   # ‚úÖ
print(int("42"))       # ‚úÖ

# Ceci provoque une ERREUR :
try:
    print(int("3.14"))  # ‚ùå int() ne peut pas convertir un str avec un point
except ValueError as e:
    print(f"‚ùå Erreur : {e}")
    print("   Solution : int(float('3.14')) =", int(float("3.14")))

try:
    print(float("bonjour"))  # ‚ùå impossible
except ValueError as e:
    print(f"\n‚ùå Erreur : {e}")
````

## 3.1 Les bool√©ens : `True` et `False`

Un **bool√©en** ne peut prendre que deux valeurs : `True` (vrai) ou `False` (faux).  
Ils sont essentiels pour les **d√©cisions** et les **filtres** en programmation.

### Applications en biologie :
- Pr√©sence / absence d'une esp√®ce
- √âchantillon valide / invalide
- Condition respect√©e / non respect√©e (pH > 7, T¬∞ > 20¬∞C...)

````python
# Cr√©ation de bool√©ens
espece_presente = True
echantillon_contamine = False

print(f"Esp√®ce pr√©sente ? {espece_presente}")
print(f"√âchantillon contamin√© ? {echantillon_contamine}")
print(f"Type : {type(espece_presente)}")

# Les bool√©ens sont aussi des nombres !
print(f"\nTrue  en nombre = {int(True)}")   # 1
print(f"False en nombre = {int(False)}")     # 0

# Utile pour compter !
presences = [True, True, False, True, False, True, True, False]
nb_presences = sum(presences)  # sum() additionne les True (=1)
print(f"\nPr√©sences : {presences}")
print(f"Nombre de sites avec pr√©sence : {nb_presences} / {len(presences)}")
````

## 3.2 Op√©rateurs de comparaison

Les comparaisons retournent toujours un **bool√©en** :

| Op√©rateur | Signification | Exemple | R√©sultat |
|-----------|--------------|---------|----------|
| `==` | √âgal √† | `5 == 5` | `True` |
| `!=` | Diff√©rent de | `5 != 3` | `True` |
| `<` | Inf√©rieur √† | `3 < 5` | `True` |
| `>` | Sup√©rieur √† | `3 > 5` | `False` |
| `<=` | Inf√©rieur ou √©gal | `5 <= 5` | `True` |
| `>=` | Sup√©rieur ou √©gal | `6 >= 5` | `True` |

> ‚ö†Ô∏è Ne confondez pas `=` (affectation) et `==` (comparaison) !

````python
# Comparaisons en contexte biologique
temperature = 24.3
pH = 7.8
salinite = 38.5
profondeur = 12

print("--- Tests de conditions environnementales ---")
print(f"Temp√©rature ({temperature}¬∞C) > 20¬∞C ? ‚Üí {temperature > 20}")
print(f"pH ({pH}) == 7.0 ?                ‚Üí {pH == 7.0}")
print(f"pH ({pH}) >= 7.0 ?                ‚Üí {pH >= 7.0}")
print(f"Salinit√© ({salinite}) != 35.0 ?    ‚Üí {salinite != 35.0}")
print(f"Profondeur ({profondeur}m) <= 10m ? ‚Üí {profondeur <= 10}")

# Comparaison de cha√Ænes (ordre alphab√©tique)
print(f"\n'Sar' == 'Sar'     ? ‚Üí {('Sar' == 'Sar')}")
print(f"'Sar' == 'sar'     ? ‚Üí {('Sar' == 'sar')}")  # sensible √† la casse !
print(f"'Algue' < 'Poisson' ? ‚Üí {('Algue' < 'Poisson')}")  # ordre alphab√©tique
````

## 3.3 Op√©rateurs bool√©ens : `and`, `or`, `not`

Ces op√©rateurs combinent plusieurs conditions :

| Op√©rateur | Description | Exemple |
|-----------|------------|----------|
| `and` | Vrai si **les deux** sont vrais | `True and False` ‚Üí `False` |
| `or` | Vrai si **au moins un** est vrai | `True or False` ‚Üí `True` |
| `not` | Inverse le bool√©en | `not True` ‚Üí `False` |

### Tables de v√©rit√© :

```
  AND          OR           NOT
 T and T = T   T or T = T   not T = F
 T and F = F   T or F = T   not F = T
 F and T = F   F or T = T
 F and F = F   F or F = F
```

````python
# Tables de v√©rit√© ‚Äî V√©rification
print("=== AND ===")
print(f"True  and True  = {True and True}")
print(f"True  and False = {True and False}")
print(f"False and True  = {False and True}")
print(f"False and False = {False and False}")

print("\n=== OR ===")
print(f"True  or True  = {True or True}")
print(f"True  or False = {True or False}")
print(f"False or True  = {False or True}")
print(f"False or False = {False or False}")

print("\n=== NOT ===")
print(f"not True  = {not True}")
print(f"not False = {not False}")
````

## 3.4 Application : Conditions environnementales combin√©es

En √©cologie, on combine souvent plusieurs crit√®res pour √©valuer un habitat :

````python
# Param√®tres mesur√©s
temperature = 21.5   # ¬∞C
pH = 8.1
salinite = 37.2      # PSU
turbidite = 3.5      # NTU
oxygene = 6.8        # mg/L

# Crit√®res pour un habitat favorable √† Posidonia oceanica
temp_ok = 15 <= temperature <= 28      # entre 15 et 28¬∞C
ph_ok = 7.5 <= pH <= 8.5               # pH l√©g√®rement basique
sal_ok = 34 <= salinite <= 39           # salinit√© marine normale
turb_ok = turbidite < 5                 # eau claire
oxy_ok = oxygene > 5                    # bien oxyg√©n√©

print("--- √âvaluation des crit√®res ---")
print(f"Temp√©rature (15-28¬∞C)  : {temperature}¬∞C  ‚Üí {'‚úÖ' if temp_ok else '‚ùå'}")
print(f"pH (7.5-8.5)           : {pH}      ‚Üí {'‚úÖ' if ph_ok else '‚ùå'}")
print(f"Salinit√© (34-39 PSU)   : {salinite} PSU ‚Üí {'‚úÖ' if sal_ok else '‚ùå'}")
print(f"Turbidit√© (< 5 NTU)    : {turbidite} NTU ‚Üí {'‚úÖ' if turb_ok else '‚ùå'}")
print(f"Oxyg√®ne (> 5 mg/L)     : {oxygene} mg/L ‚Üí {'‚úÖ' if oxy_ok else '‚ùå'}")

# Combinaison avec AND
habitat_favorable = temp_ok and ph_ok and sal_ok and turb_ok and oxy_ok
print(f"\nüåø Habitat favorable ? ‚Üí {habitat_favorable}")

# Au moins un crit√®re probl√©matique ?
probleme = not temp_ok or not ph_ok or not sal_ok or not turb_ok or not oxy_ok
print(f"‚ö†Ô∏è  Au moins un probl√®me ? ‚Üí {probleme}")
````
