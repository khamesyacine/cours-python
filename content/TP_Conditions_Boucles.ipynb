# Partie 1 ‚Äî Conditions : `if` / `elif` / `else`
---

Les conditions permettent au programme de **prendre des d√©cisions**.

```python
if condition:
    # ex√©cut√© si condition est True
elif autre_condition:
    # ex√©cut√© si la premi√®re est False et celle-ci est True
else:
    # ex√©cut√© si toutes les conditions sont False
```

> ‚ö†Ô∏è N'oubliez pas les **deux points `:`** et l'**indentation** (4 espaces).

## 1.1 `if` / `else` simple

````python
glycemie = 1.45                    # g/L √† jeun
if glycemie > 1.26:
    print("Diab√®te")
else:
    print("Pas de diab√®te")
````

````python
temperature = 38.9                 # ¬∞C
if temperature >= 38.0:
    print("Fi√®vre")
else:
    print("Temp√©rature normale")
````

````python
hemoglobine = 10.2                 # g/dL
if hemoglobine < 12:
    print("An√©mie")
else:
    print("H√©moglobine normale")
````

## 1.2 `if` / `elif` / `else` ‚Äî Plusieurs cas

````python
# Classification de la glyc√©mie √† jeun
glycemie = 1.15
if glycemie < 0.70:
    print("Hypoglyc√©mie")
elif glycemie <= 1.10:
    print("Normal")
elif glycemie <= 1.26:
    print("Pr√©-diab√®te")
else:
    print("Diab√®te")
````

````python
# Classification de l'IMC
IMC = 27.3
if IMC < 18.5:
    print("Insuffisance pond√©rale")
elif IMC < 25:
    print("Poids normal")
elif IMC < 30:
    print("Surpoids")
else:
    print("Ob√©sit√©")
````

````python
# Interpr√©tation du pH sanguin
pH = 7.30
if pH < 7.35:
    print("Acidose")
elif pH > 7.45:
    print("Alcalose")
else:
    print("pH normal")
````

````python
# R√©sistance aux antibiotiques (CMI)
CMI = 24
if CMI <= 4:
    print("Sensible")
elif CMI <= 16:
    print("Interm√©diaire")
else:
    print("R√©sistant")
````

````python
# Groupe sanguin et transfusion
groupe = "O-"
if groupe == "O-":
    print("Donneur universel")
elif groupe == "AB+":
    print("Receveur universel")
else:
    print("Compatibilit√© √† v√©rifier")
````

## 1.3 Conditions combin√©es avec `and` / `or`

````python
# Carence en fer : h√©moglobine basse ET ferritine basse
hemoglobine = 10.5
ferritine = 8.0
if hemoglobine < 12 and ferritine < 20:
    print("Carence en fer confirm√©e")
else:
    print("Pas de carence en fer")
````

````python
# Patient √† risque cardiovasculaire
fumeur = True
diabetique = True
age = 55
if (fumeur and diabetique) or age > 60:
    print("Risque cardiovasculaire √©lev√©")
else:
    print("Risque mod√©r√©")
````

# Partie 2 ‚Äî Boucle `for`
---

La boucle `for` **parcourt** chaque √©l√©ment d'une liste, un par un.

```python
for element in liste:
    # action sur chaque element
```

## 2.1 Parcourir une liste

````python
patients = ["Ali", "Sara", "Karim", "Nadia"]
for p in patients:
    print(f"Patient : {p}")
````

````python
glycemies = [0.95, 1.32, 0.88, 1.50, 1.05]
for g in glycemies:
    print(f"Glyc√©mie : {g} g/L")
````

````python
# Compter les bases d'une s√©quence ADN
sequence = "ATCGGATCGA"
compteur_A = 0
for base in sequence:
    if base == "A":
        compteur_A += 1
print(f"Nombre de A : {compteur_A}")
````

## 2.2 `for` + `if` ‚Äî Filtrer des donn√©es

````python
# Identifier les glyc√©mies anormales
glycemies = [0.95, 1.32, 0.88, 1.50, 1.05, 0.72, 1.28]
for g in glycemies:
    if g > 1.26:
        print(f"{g} g/L ‚Üí Diab√®te")
````

````python
# Compter les patients f√©briles
temperatures = [37.2, 38.5, 36.9, 39.1, 37.0, 38.8]
febriles = 0
for t in temperatures:
    if t >= 38.0:
        febriles += 1
print(f"Patients f√©briles : {febriles} / {len(temperatures)}")
````

````python
# Identifier les souches r√©sistantes
CMIs = [2, 8, 32, 4, 64, 16, 128]
for cmi in CMIs:
    if cmi > 16:
        print(f"CMI = {cmi} ‚Üí R√©sistant")
    else:
        print(f"CMI = {cmi} ‚Üí Sensible")
````

````python
# Classer les IMC d'un groupe de patients
IMCs = [22.1, 27.5, 31.0, 18.2, 24.8, 35.6]
for imc in IMCs:
    if imc < 18.5:
        print(f"IMC {imc} ‚Üí Maigreur")
    elif imc < 25:
        print(f"IMC {imc} ‚Üí Normal")
    elif imc < 30:
        print(f"IMC {imc} ‚Üí Surpoids")
    else:
        print(f"IMC {imc} ‚Üí Ob√©sit√©")
````

# Partie 3 ‚Äî Boucle `while`
---

La boucle `while` **r√©p√®te** tant qu'une condition est vraie.

```python
while condition:
    # action r√©p√©t√©e
```

> ‚ö†Ô∏è Assurez-vous que la condition devienne `False` √† un moment, sinon la boucle ne s'arr√™te jamais !

## 3.1 Exemples de base

````python
# Croissance bact√©rienne jusqu'√† 1 million
bacteries = 1
generations = 0
while bacteries < 1000000:
    bacteries = bacteries * 2
    generations += 1
print(f"1 million atteint en {generations} g√©n√©rations")
````

````python
# √âlimination d'un m√©dicament : combien d'heures pour passer sous 10 mg ?
dose = 500.0
heures = 0
while dose > 10:
    dose = dose * 0.7              # 30% √©limin√© par heure
    heures += 1
print(f"Sous 10 mg apr√®s {heures} heures")
````

````python
# Diluer jusqu'√† concentration < 0.1 mg/L
concentration = 50.0
dilutions = 0
while concentration >= 0.1:
    concentration = concentration / 2
    dilutions += 1
print(f"Il faut {dilutions} dilutions au 1/2")
````

````python
# R√©plication ADN : combien de cycles PCR pour 1 milliard de copies ?
copies = 1
cycles = 0
while copies < 1000000000:
    copies = copies * 2
    cycles += 1
print(f"1 milliard de copies en {cycles} cycles PCR")
````

````python
# D√©croissance radioactive (demi-vie)
activite = 1000.0                  # Becquerels
demi_vies = 0
while activite > 50:
    activite = activite / 2
    demi_vies += 1
print(f"Activit√© < 50 Bq apr√®s {demi_vies} demi-vies")
````

````python
# Perte de poids : combien de semaines pour atteindre 75 kg ?
poids = 95.0
objectif = 75.0
semaines = 0
while poids > objectif:
    poids = poids - 0.5            # perte de 0.5 kg/semaine
    semaines += 1
print(f"Objectif atteint en {semaines} semaines")
````

## 3.2 `while` avec un compteur

````python
# Prises de m√©dicament : 3 fois par jour pendant 7 jours
jour = 1
while jour <= 7:
    print(f"Jour {jour} : 3 prises")
    jour += 1
````

````python
# Suivi de glyc√©mie sur 5 jours
jour = 1
glycemie = 1.80                    # d√©part √©lev√©
while jour <= 5:
    print(f"Jour {jour} : glyc√©mie = {glycemie:.2f} g/L")
    glycemie = glycemie - 0.15     # am√©lioration sous traitement
    jour += 1
````
